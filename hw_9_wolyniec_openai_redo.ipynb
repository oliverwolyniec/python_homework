Step 3: Obtain Embeddings for Key Phrases
Get embeddings for each key phrase. These embeddings will serve as our reference for identifying similar content in the document.

```python
def get_embeddings(text):
    response = openai.Embedding.create(input=text, model="text-embedding-ada-002") #There are different embeddings you can choose. see documentation
    return np.array(response['data'][0]['embedding'])

# Generate embeddings for each key phrase
key_embeddings = {phrase: get_embeddings(phrase) for phrase in key_phrases}
```

Step 4: Process Document and Split Text
Break the document text into smaller segments (e.g., by paragraphs or sentences) to make the search more accurate.

```python
import fitz
#Tried a few different ways to extract text from the PDF and landed on this one, other ways weren't extracting properly but I tried a few different fixes with ChatGPT
pdf_path = r"C:\Users\wolyn\Downloads\bennyyyy.pdf" #I trimmed the PDF so as to not exceed the token limit
document_text = ""
with fitz.open(pdf_path) as pdf:
    for page in pdf:
        document_text += page.get_text()

lines = document_text.split('\n')

paragraphs = []
current_paragraph = ""

for line in lines:
    line = line.strip()
    if line:
        if current_paragraph and not current_paragraph.endswith(('.', '!', '?')): #The PDF wasn't being processed properly (was producing at times single extremely long strings without spaces, then inserting odd spaces in the middle of words at other points, so I opted to troubleshoot with ChatGPT)
            current_paragraph += " " + line
        else:
            if current_paragraph:
                paragraphs.append(current_paragraph)
            current_paragraph = line

if current_paragraph:
    paragraphs.append(current_paragraph)

for i, paragraph in enumerate(paragraphs): #Checking on the paragraphs here to make sure the way they're divvied up makes sense and can be analyzed easily
    print(f"Paragraph {i+1}: {paragraph}")
```

Step 5: Get Embeddings for Each Paragraph and Find Similarities
Compute embeddings for each paragraph and find the similarity between each paragraph’s embedding and the key phrase embeddings. Cosine similarity can be used for this.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Get embeddings for each paragraph
paragraph_embeddings = [get_embeddings(paragraph) for paragraph in paragraphs]

# Function to find similar paragraphs based on cosine similarity
def find_similar_paragraphs(paragraph_embeddings, key_embeddings, threshold=0.78):
    similar_paragraphs = []
    
    for i, paragraph_emb in enumerate(paragraph_embeddings):
        for phrase, key_emb in key_embeddings.items():
            similarity = cosine_similarity([paragraph_emb], [key_emb])[0][0]
            if similarity > threshold:
                similar_paragraphs.append((paragraphs[i], phrase, similarity))
                
    return similar_paragraphs

# Find paragraphs similar to the key phrases
results = find_similar_paragraphs(paragraph_embeddings, key_embeddings)
#Played around with the threshold and landed on 0.78, which gave me a good number of key paragraphs to work with.
#Interesting that a seemingly small change (from 0.8 to 0.75) seemed to disproportionately widen the search and gave me tons and tons of similar paragraphs
```

Step 6: Display Results
Print the paragraphs that are most similar to our key phrases, along with the similarity score.

```python
for paragraph, phrase, similarity in results:
    print(f"Similarity Score: {similarity:.2f}")
    print(f"Matched Key Phrase: {phrase}")
    print("Paragraph:", paragraph)
    print("-" * 80)
```

The code above uses vector-based similarity to locate passages in the document that closely align with our key phrases. Adjust the threshold parameter to control the strictness of the match; a higher threshold (e.g., 0.85) will yield more specific matches, while a lower threshold (e.g., 0.7) will yield broader matches.

```python
import openai
import fitz  # PyMuPDF for PDF text extraction

# Set up your OpenAI API key (replace with your own key)
openai.api_key = "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"  # Replace with your OpenAI API key

# Function to extract text from PDF
def extract_pdf_text(pdf_path):
    with fitz.open(pdf_path) as pdf_document:
        text = ""
        for page_num in range(pdf_document.page_count):
            page = pdf_document.load_page(page_num)
            text += page.get_text("text")  # Extract text from each page
        return text

# Function to chunk text if it's too long (helpful for large documents)
def chunk_text(text, max_chunk_size=2000):
    """Splits the text into chunks if it's too long."""
    chunks = []
    while len(text) > max_chunk_size:
        # Find the last period before max_chunk_size to break the chunk
        split_point = text.rfind('.', 0, max_chunk_size)
        chunks.append(text[:split_point + 1])
        text = text[split_point + 1:].lstrip()  # Remove the chunk we just added
    chunks.append(text)  # Add the remaining part as the last chunk
    return chunks

# Function to query ChatGPT with the extracted text
def query_chatgpt_with_pdf_content(document_text, question):
    # Chunk the document text if it is too long
    text_chunks = chunk_text(document_text)
    
    answers = []
    for chunk in text_chunks:
        prompt = f"{chunk}\n\nQuestion: {question}\nAnswer:"
        
        # Query ChatGPT for each chunk of text
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",  # You can switch to "gpt-4" if you have access
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.5,
            max_tokens=150  # Adjust max tokens as needed
        )
        
        # Extract the response from OpenAI's API and append to answers list
        answer = response['choices'][0]['message']['content'].strip()
        answers.append(answer)
    
    # Combine the answers from all chunks and return
    return " ".join(answers)

# Path to your PDF
pdf_path = r"C:\Users\wolyn\Downloads\anthro.pdf"  # Replace with your PDF path

# Extract text from PDF
pdf_content = extract_pdf_text(pdf_path)

# Ask a question based on the document's content
question_from_pdf = "What does the document talk about?"  # Example question related to the PDF content

# Query ChatGPT with the extracted document text and your question
answer_from_pdf = query_chatgpt_with_pdf_content(pdf_content, question_from_pdf)

# Print the answer related to the document
print(f"Answer to your question (from PDF content): {answer_from_pdf}")
```
